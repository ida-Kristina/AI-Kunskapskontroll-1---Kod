Den här filen gjorde jag för att kunna utforska hur datan ser ut när den komemr från min API. Jag har tittat på vilka nycklar som finns, samt hur datan är strukturerad. Jag kom fram till att jag kunde använda mig av den inbygda html-strukturen (ul, li h# etc.) som en grund för att bygga mina chunks. Jag ser också att syfte är relativt kortfattat, medan kunskapskrav och centralt innehåll är mycket omfattande och kommer behöva brytas ner. Det känns också sannolikt som att det kommer vara viktigt att filtrera baserat på åk för att svara på användarens frågor.

```python
import json
from pprint import pprint

with open("matematik.json", "r", encoding="utf-8") as f:
    data = json.load(f)

subject = data["subject"]

for key, value in subject.items():
    print("\n---", key, "---")
    print(type(value))

    if isinstance(value, list):
        print("Antal objekt:", len(value))
        if len(value) > 0:
            pprint(value[0])

    elif isinstance(value, dict):
        pprint(value)

    else:
        print(value)
```

```python
import json
from pprint import pprint

with open("matematik.json", "r", encoding="utf-8") as f:
    data = json.load(f)

subject = data["subject"]

print("centralContents finns:", "centralContents" in subject)
print("knowledgeRequirements finns:", "knowledgeRequirements" in subject)

print("\nFörsta centralContents-objektet:")
pprint(subject["centralContents"][0], width=120)

print("\nFörsta knowledgeRequirements-objektet:")
pprint(subject["knowledgeRequirements"][0], width=120)
```
